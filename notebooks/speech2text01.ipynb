## NLP App - Speech 2 Text



```python
audio_file = 'speech2text01.wav' #Read from wikipedia definition of Speech Recognition
# audio_file = 'speech2text_med03.wav'
```

```python
from scipy.io.wavfile import read, write
from IPython.display import Audio

Fs, data = read(audio_file)

data = data[:,0]

Audio(data, rate=Fs)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(data)
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.title('Waveform of: ' + audio_file)
plt.show()
```

```python
from speechbrain.pretrained import EncoderDecoderASR

asr_model = EncoderDecoderASR.from_hparams(source="speechbrain/asr-crdnn-rnnlm-librispeech", savedir="pretrained_model")
# asr_model = EncoderDecoderASR.from_hparams(source="speechbrain/mtl-mimic-voicebank", savedir="pretrained_model")

asr_model.transcribe_file(audio_file)
```
